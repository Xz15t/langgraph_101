```python
from langchain_openai import ChatOpenAI

import os
from dotenv import load_dotenv

load_dotenv()
if os.environ.get("OPENAI_API_KEY"):
    print("Open AI API Key is set.")
else:
    raise ValueError("Open AI API Key is not set")
```

```python
llm = ChatOpenAI(model = "gpt-5-nano")
llm
```

# **First Graph**

### **Step1: Define the schema**

```python
# schema should not change
from typing import TypedDict, List

class graph_schema(TypedDict):
    name: str
    message: str
```

### **Step2: Create the node function***

```python
def welcome(state:graph_schema) -> graph_schema:
    curr_name = state['name']
    curr_msg = state['message']

    response = llm.invoke(f"My name is {curr_name}.{curr_msg}").content

    # in-place changes in dictionary (Pointers)
    state['message'] = f"Your message was {curr_msg}. Here is my response : {response}"

    return state
```

### **Step3: Create the state graph**

```python
from langgraph.graph import StateGraph, START, END

graph =  StateGraph(graph_schema)

# Adding Nodes
graph.add_node("welcome",welcome) # name of node and what it does

# Adding Edges
graph.add_edge(START, "welcome")
graph.add_edge("welcome", END)
```

### **Step4: Compile my graph**

```python
from IPython.display import Image, display

first_graph = graph.compile()

# If PNG shows error
Image(first_graph.get_graph().draw_mermaid_png())

# Code to make graph online
print(first_graph.get_graph().draw_mermaid())
```

### **Step5: Run my graph**

```python
first_graph.invoke({"name":"Mukund","message":"How are you?"})
```
